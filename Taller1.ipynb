```python
def categorizar_sitio(sitio):
    sitio = sitio.lower()

    # Supermercados, tiendas y alimentos
    if 'supermercado' in sitio or 'tienda express' in sitio or 'almacen por departamento' in sitio:
        return 'Supermercados y Alimentos'

    # Vestuario y accesorios
    elif 'vestuario' in sitio or 'calzado' in sitio or 'outlet' in sitio:
        return 'Vestuario y Accesorios'

    # Compras en internet (se asume que este patrón es claro)
    elif 'compras en internet' in sitio:
        return 'Compras en internet'

    # Salud y Bienestar
    elif ('drogueria' in sitio or 'farmacia' in sitio or 'tienda naturista' in sitio or
          'optica' in sitio or 'clinica' in sitio or 'servicios medicos' in sitio or
          'medicina' in sitio):
        return 'Salud y Bienestar'

    # Viajes y Transporte
    elif ('aerolinea' in sitio or 'viaje' in sitio or 'hotel' in sitio or
          'agencia de viaje' in sitio or 'aviacion' in sitio or 'motel' in sitio or
          'estacion de servicio' in sitio or 'vehiculo' in sitio or 'llanta' in sitio or
          'mecanica' in sitio):
        return 'Viajes y Transporte'

    # Entretenimiento y Cultura
    elif ('casino' in sitio or 'club social' in sitio or 'boleteria' in sitio or
          'sala de cine' in sitio or 'teatro' in sitio or 'libreria' in sitio or
          'disco' in sitio or 'periodico' in sitio or 'revista' in sitio or
          'musica' in sitio):
        return 'Entretenimiento y Cultura'

    # Hogar y Construcción
    elif ('mueble' in sitio or 'almacen de articulos para el hogar' in sitio or
          'ferreteria' in sitio or 'materiales de construccion' in sitio or
          'vidrio' in sitio or 'alfombra' in sitio or 'piso' in sitio):
        return 'Hogar y Construcción'

    # Tecnología y Telecomunicaciones
    elif ('computador' in sitio or 'equipo' in sitio or 'tecnologia' in sitio or
          'telecomunicacion' in sitio or 'software' in sitio):
        return 'Tecnología y Telecomunicaciones'

    # Financiero y Seguros
    elif ('aseguradora' in sitio or 'pago de impuestos' in sitio or
          'servicios de seguros' in sitio):
        return 'Financiero y Seguros'

    # Educación y Servicios Profesionales
    elif ('colegios' in sitio or 'universidades' in sitio or 'institutos' in sitio or
          'asociacion' in sitio or 'gremio' in sitio):
        return 'Educación y Servicios Profesionales'

    # Si no se encontró coincidencia, se agrupa en Otros
    else:
        return 'Otros'
```

##Modeling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#Categorical variables are converted to integers to ensure proper processing
data_cleaned_vf=pd.get_dummies(data_cleaned, dtype='int64')
data_cleaned_vf.head()
```

###Dimensionality Reduction

```python
#Data Normalization
sc=StandardScaler()
data_cleaned_vf_sc=sc.fit_transform(data_cleaned_vf)
data_cleaned_vf_sc
```

###PCA

```python
pca=PCA(n_components=15)
data_cleaned_vf_pca=pca.fit_transform(data_cleaned_vf_sc)
print(f" Varianza por componente: {pca.explained_variance_ratio_}")
#Al hacer la varianza acumulada entonces se puede entender cuanto del total de la varianza se tiene con todos los componentes
print(f"Varianza acumulada: {pca.explained_variance_ratio_.cumsum()}")
```

###K-Means

####Técnica del codo (inercia)

```python
inercia=[]
for i in range(2,12):
  kmeans=KMeans(n_clusters=i, random_state=42)
  kmeans.fit(data_cleaned_vf_pca)
  inercia.append(kmeans.inertia_)
plt.plot(range(2,12),inercia, marker="x")
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.show()
```

####Silhouette

```python
silhouette_scores = []
for i in range(2, 12):
    kmeans2 = KMeans(n_clusters=i, random_state=42)
    kmeans2_labels = kmeans2.fit_predict(data_cleaned_vf_pca)
    score=silhouette_score(data_cleaned_vf_pca, kmeans2_labels)
    silhouette_scores.append(score)
plt.plot(range(2,12),silhouette_scores, marker="x")
plt.xlabel("Número de clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

###Results

```python
grupos=KMeans(n_clusters=4, random_state=42)
grupos.fit(data_cleaned_vf_pca)
print(grupos.labels_)
```
